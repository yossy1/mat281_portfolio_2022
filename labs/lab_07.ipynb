El conjunto de datos se denomina `creditcard.csv` y consta de varias columnas con información acerca del fraude de tarjetas de crédito, en donde la columna **Class** corresponde a: 0 si no es un fraude y 1 si es un fraude.

En este ejercicio se trabajará el problemas de  clases desbalancedas. Veamos las primeras cinco filas dle conjunto de datos:

```python
#Importar librerias, modelos y méticas.
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn import datasets
from sklearn.model_selection import train_test_split

from sklearn.metrics import confusion_matrix,accuracy_score,recall_score,precision_score,f1_score
from sklearn.dummy import DummyClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier

%matplotlib inline
sns.set_palette("deep", desat=.6)
sns.set(rc={'figure.figsize':(11.7,8.27)})
```

```python
# cargar datos
df = pd.read_csv("https://raw.githubusercontent.com/fralfaro/MAT281_2022/main/labs/lab_07/data/creditcard.csv",sep=';')
df.head()
```

Analicemos el total de fraudes respecto a los casos que nos son fraudes:


```python
#DataFrame con cantidades y proporciones de fraudes y no fraudes.
#Crear Dataframe df_count vacio.
df_count = pd.DataFrame()
#Definir columnas del DataFrame df_count.
df_count["fraude"] =["no","si"]
df_count["total"] = df["Class"].value_counts() 
df_count["porcentaje"] = 100*df_count["total"] /df_count["total"] .sum()
# Mostrar df_count
df_count
```

Se observa que menos del 1% corresponde a registros frudulentos. La pregunta que surgen son:

* ¿ Cómo deben ser el conjunto de entrenamiento y de testeo?
* ¿ Qué modelos ocupar?
* ¿ Qué métricas ocupar?

Por ejemplo, analicemos el modelos de regresión logística y apliquemos el procedimiento estándar:

```python
# datos 
y = df.Class
X = df.drop('Class', axis=1)
# split o separar dataset en conjuntos test y entrenamiento.
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=27)
# Creando el modelo y ajustando el modelo a los datos de entrenamiento.
lr = LogisticRegression(solver='liblinear').fit(X_train, y_train) 
# predecir 
lr_pred = lr.predict(X_test)
# calcular accuracy
accuracy_score(y_test, lr_pred)
```

En general el modelo tiene un **accuracy** del 99,9%, es decir, un podría suponer que el modelo predice casi perfectamente, pero eso esta lejos de ser así.  Para ver por qué es necesario seguir los siguientes pasos:

### 1. Cambiar la métrica de rendimiento

El primer paso es comparar con distintas métricas, para eso ocupemos las 4 métricas clásicas abordadas en el curso:
* accuracy
* precision
* recall
* f-score

En este punto deberá poner las métricas correspondientes y comentar sus resultados.

```python
# metrics
#Lista con datos de testeo.
y_true_1 =  list(y_test)
#Lista con datos que predice el modelo.
y_pred_1 = list(lr.predict(X_test))
#Matriz de confusión.
print('\nMatriz de confusion:\n ')
print(confusion_matrix(y_true_1,y_pred_1))
print('\nMetricas:\n ')
# Aplicar distintas métricas a los datos de testeo y que predijo el modelo.
print('accuracy:   ',accuracy_score(y_test, lr_pred))
print('recall:     ',recall_score(y_test, lr_pred))
print('precision:  ',precision_score(y_test, lr_pred))
print('f-score:    ',f1_score(y_test, lr_pred))
print("")# Imprimir resultados.
```

```python
# Crear DataFrame df_result_1 con las métricas de LogisticRegression. 
#DataFrame vacio.
df_result_1 = pd.DataFrame()
#Definir columnas del DataFrame.
df_result_1['accuracy'] = [accuracy_score(y_true_1,y_pred_1)]
df_result_1['recall'] = [recall_score(y_true_1,y_pred_1)]
df_result_1['precision'] = [precision_score(y_true_1,y_pred_1)]
df_result_1['fscore'] = [f1_score(y_true_1,y_pred_1)]
df_result_1['modelo'] = ['LogisticRegression']
df_result_1#Mostrar DataFrame.
```

Según las métricas podemos concluir que con este modelo se consigue: 
* (accuracy) Un porcentaje casi total de predicciones correctas frente al total de predicciones.
* (recall) Una proporción apenas aceptable de casos positivos bien clasificados por el modelo, respecto al total de positivos, o mejor dicho,  una habilidad aceptable del modelo para detectar los casos relevantes.
* (precision) La mayoria de resultado de predicciones cercanos al valor verdadero.
* (f-score) Una buena precisión de test.


### 2. Cambiar algoritmo

El segundo paso es comparar con distintos modelos. Debe tener en cuenta que el modelo ocupaod resuelva el problema supervisado de clasificación.

En este punto deberá ajustar un modelo de **random forest**, aplicar las métricas y comparar con el modelo de regresión logística.

```python
#Entrenar modelo Random Forest.
rfc = RandomForestClassifier(max_depth=None, n_estimators=100, max_features="sqrt").fit(X_train, y_train) # algoritmo random forest
```

```python
# Crear DataFrame df_result_2 con las métricas de RandomForest. 
#DataFrame vacio.
df_result_2 = pd.DataFrame()
#Definir columnas del DataFrame.
df_result_2['accuracy'] = [accuracy_score(y_true_2,y_pred_2)]
df_result_2['recall'] = [recall_score(y_true_2,y_pred_2)]
df_result_2['precision'] = [precision_score(y_true_2,y_pred_2)]
df_result_2['fscore'] = [f1_score(y_true_2,y_pred_2)]
df_result_2['modelo'] = ['RandomForest']
df_result_2#Mostrar DataFrame.
```

Según estas métricas podemos ver que, a diferencia del anterior modelo, con este modelo se consigue:

### 3. Técnicas de remuestreo: sobremuestreo de clase minoritaria

El tercer paso es ocupar ténicas de remuestreo, pero sobre la clase minoritaria. Esto significa que mediantes ténicas de remuestreo trataremos de equiparar el número de elementos de la clase minoritaria a la clase mayoritaria.

```python
from sklearn.utils import resample
# concatenar el conjunto de entrenamiento
X = pd.concat([X_train, y_train], axis=1)
# separar las clases
not_fraud = X[X.Class==0]
fraud = X[X.Class==1]
# remuestrear  clase minoritaria
fraud_upsampled = resample(fraud,
                          replace=True, # sample with replacement
                          n_samples=len(not_fraud), # match number in majority class
                          random_state=27) # reproducible results
# recombinar resultados
upsampled = pd.concat([not_fraud, fraud_upsampled])
# chequear el número de elementos por clases
upsampled.Class.value_counts()# Cantidad de elementos por clase.
```

```python
# datos de entrenamiento sobre-balanceados
y_train = upsampled.Class
X_train = upsampled.drop('Class', axis=1)
```

Ocupando estos nuevos conjunto de entrenamientos, vuelva a aplicar el modelos de regresión logística y calcule las correspondientes métricas. Además, justifique las ventajas y desventjas de este procedimiento.

```python
#Entrenar modelo con nuevos datos.
upsampled = LogisticRegression(solver='liblinear').fit(X_train, y_train) # algoritmo de regresion logistica
# metrics
#Lista con datos de testeo.
y_true_3 =  list(y_test)
#Lista con datos que predice el modelo.
y_pred_3 = list(upsampled.predict(X_test))
#Matriz de confusión.
print('\nMatriz de confusion:\n ')
print(confusion_matrix(y_true_3,y_pred_3))
# Aplicar distintas métricas a los datos de testeo y que predijo el modelo.
print('\nMetricas:\n ')
print('accuracy:   ',accuracy_score(y_true_3,y_pred_3))
print('recall:     ',recall_score(y_true_3,y_pred_3))
print('precision:  ',precision_score(y_true_3,y_pred_3))
print('f-score:    ',f1_score(y_true_3,y_pred_3))
print("")# Imprimir resultados.
```

```python
# Crear DataFrame df_result_3 con las métricas de LogisticRegression Sobre-balanceado. 
#DataFrame vacio.
df_result_3 = pd.DataFrame()
#Definir columnas del DataFrame.
df_result_3['accuracy'] = [accuracy_score(y_true_3,y_pred_3)]
df_result_3['recall'] = [recall_score(y_true_3,y_pred_3)]
df_result_3['precision'] = [precision_score(y_true_3,y_pred_3)]
df_result_3['fscore'] = [f1_score(y_true_3,y_pred_3)]
df_result_3['modelo'] = ['LogisticRegression_Sobre-balanceado']
df_result_3#Mostrar DataFrame.
```

Según las métricas de este modelo agustado podemos comparar que: 
* (accuracy o exactitud) Bajo en 1,98% el total de predicciones correctas frente al total de predicciones.
* (recall o sensibilidad) Subio en 15,44% la proporción de casos positivos bien clasificados por el modelo, respecto al total de positivos, o mejor dicho, la habilidad del modelo para detectar los casos relevantes.
* (precision o presiciòn) Bajo en 56,38% los resultado de predicciones cercanos al valor verdadero.
* (f-score) Bajo en 35,45% la precisión de test.

Entonces, se puede ver que sobre-abalancear los datos mejorò la sensibilidad y que la exactitud bajo infimamente. Sin embargo, se perdió precisión en gran medida y f_score empeoró.
